### Problem 1 - M/L ratios and IMF

 In the end of the ex03 notebook I used some examples of the stellar population synthesis calculations, which assumed Chabrier initial mass function (IMF) of stars to estimate $M_\star/L_r\approx 2$ for the Milky Way's color. Would the estimated $M_\star/L_r$ be larger or smaller, if the Salpeter IMF was assumed instead? Present an argument for your answer based on what we discussed about IMFs in class (see discussion and IMF related plots earlier in ex03 notebook to aid your answer).


### Solution

In the case of Chabrier IMF, the mass function per log interval   ($ dn / d\ln m $)  is described by a log-normal form for low masses ($ m \gtrsim 0.1 M_\odot $) with a peak mass $m_p$ at masses $ m \lesssim 1 M_\odot$, while at larger masses can be described as a power law, $dn / d\ln m \propto m^{-x} $ with $ x \approx 1.35$.

On the other hand the Salpeter IMF is a power law for the whole range of masses.

Thus, the number of low mass stars is suppressed in the case of Chabrier IMF in comparison to the case of Salpeter IMF. Salpeter IMF gives higher number of massive stars than Chabries IMF. 

For mid-to-high mass stars, the mass-luminosity relation is:

\begin{equation}
L \propto M^{3.5 - 4}
\end{equation}

Thus higher mass stars have a mass to luminosity ratio: $M/L \propto M^{-(2.5 - 3)}$. So, the more massive the star, the lower its mass to luminosity ratio is. 

Since the Salpeter IMF gives a lower number of massive stars compared to the Chabrier, we expect from the above argument that the estimated $M_\star / L_r$ would be smaller if the Salpeter IMF was assumed insted of the Chabrier IMP.





### Problem 2 - Dark Matter halo of the Milky Way

For reasons that we will discuss a bit later in the course, dark matter halos are expected to have radial density profiles (density as a function of radial distance to the center) of the so-called Navarro-Frenk-White (NFW) form (Navarro et al. 1997):
$$\rho_{\rm dm}(r)=\frac{4\rho_s}{(r/r_s)(1+r/r_s)^2},$$
where $r_s$ is the scale radius, which corresponds to $r$ where logarithmic slope of the profile is -2, $\rho_s$ is density at $r=r_s$.

### a)
Derive the analytic expression for the mass profile of a spherical NFW halo: $$M(r)=4\pi\int\limits_0^r\rho(r)r^2dr$$ and express its normalization constant in terms of the mass within $r=250$ kpc, $M_{\rm h}(250\ \rm kpc)$. Use this expression to get the expression for the square of the circular velocity due to the halo, $V_{\rm circ}^2(r)$, and add it to the circular velocity due to the disk and spherical bulge (see script from ex04 below), so that you get the total circular velocity profile due to all three components (bulge, disk, dark matter).
### b) 
Plot the observed rotation measurements of stars and gas in the Milky Way, as shown in ex04 and script below. Find approximate value $M_{\rm h}(250\ \rm kpc)$ for which $V_{\rm circ}^2(r)$ describes observed rotation measurements well. ### Solution 

#### a)   

Using the NFW profile, we have for the mass of the $\mathbf{halo}$:

\begin{equation}
\displaystyle
M(r) = M(r)= 4\pi\int\limits_0^r\rho_{dm}(r) r^2dr = 16 \pi \rho_s r_s \int_0^r \frac{r'}{\left(1 + \frac{r'}{r_s} \right)^2} dr'
\end{equation}

Changing variables:
\begin{equation}
u = 1 + \frac{r'}{r_s} \Rightarrow r' = (u - 1)r_s, \quad \mbox{lims:} \,(1, 1 + r/r_s), \quad dr' = r_s du,
\end{equation}

the above integral becomes:
\begin{equation}
\displaystyle
M_h(r) = 16\pi \rho_s r_s^3 \int_1^{1+r/r_s} \, \frac{(u-1)}{u^2}du = \int_1^{1 + r/r_s} \left(\frac{1}{u} - \frac{1}{u^2} \right) du \,\,\Rightarrow 
\end{equation}


\begin{equation}
\Rightarrow M_h(r) = 16 \pi \rho_s r_s^3 \left(\ln \left[1+\frac{r}{r_s}\right] + \frac{r}{r_s} - 1  \right) = 16 \pi \rho_s r_s^3 \left(\ln \left[1+\frac{r}{r_s}\right]  - \frac{r}{r_s + r} \right).
\end{equation}

We want now to express the normalization constant in terms of the mass inside $250$ kpc, $M_h(250 \mbox{kpc}) \equiv M_{250}$. From the above formula, and using $r_250 \equiv 250$kpc, we see that:

\begin{equation}
M_{250} = 16 \pi \rho_s r_s^3 \left(\ln \left[1+\frac{r_{250}}{r_s}\right]  - \frac{r_{250}}{r_s + r_{250}} \right) \Rightarrow 16 \pi \rho_s r_s^3 = \alpha M_{250},
\end{equation}
where I have defined:
\begin{equation}
\alpha \equiv \frac{1}{\ln \left[1+\frac{r_{250}}{r_s}\right]  - \frac{r_{250}}{r_s + r_{250}}}.
\end{equation}


Thus we can write:

\begin{equation}
\boxed{
M_h(r) = \alpha M_{250} \left(\ln \left[1+\frac{r}{r_s}\right] - \frac{r}{r_s + r}   \right) }.
\end{equation}

Now, we can calculate the rotational velocity due to the dark matter halo. Since the halo has spherical symmetry:

\begin{equation}
V_{circ, h}^2(r) = \frac{GM(r)}{r} = \frac{G \alpha M_{250} }{r} \left(\ln \left[1+\frac{r}{r_s}\right] - \frac{r}{r_s + r}   \right).
\end{equation}

Since the contributions of different mass components to the total circular velocity should be added in quadrature, we have:

\begin{equation}
V_{circ, tot}(r) = \sqrt{ V^2_{circ, h}(r) +  V^2_{circ, b}(r) + V^2_{circ, d}(r)} 
\end{equation}

where $V_{circ, h},V_{circ, b}, V_{circ, d}$ the rotational velocities due to halo, bulge and disk, respectively.

We have presented the formula for the velocitiy due to halo. The other two components are given by:

\begin{equation}
V^2_{circ, b}(r) = \frac{GM_b r}{(r + r_b)^2},
\end{equation}
where we have used the \textit{Hernquist} mass profile and we have assumed spherically symmetric bulge (M_b, r_b char mass and radius of bulge). 


And for the disk:
\begin{equation}
V^2_{\rm circ,d}(r)= 4\pi G\Sigma_0 R_d y^2 [I_0(y)K_0(y)-I_1(y)K_1(y)]
\end{equation}


where $I_n$ and $K_n$ are the modified Bessel functions of the first and second kind, respectively, $r_{\rm d}$ is the disk scale length, and $y=r/(2r_{\rm d})$. The total mass of the exponential disk is $M_{\rm d}=2\pi\Sigma_0 r_{\rm d}^2$.

We can can then plot the rotation curve, using the above formulae. We use different values for the mass within 250 kpc, but I use the knowledge that the mass of DM halo is $\sim 10^{12} \, M_\odot$, to make less trials.

### Problem 4 - Modern estimate of the MW+M31 mass using the Kahn-Woltjer method

Summarize the argument behind the Kahn-Woltjer (1959) method of estimating the sum of total gravitating masses of the Milky Way and M31 galaxies discussed in the ex04 notebook and S 1.9.3 in the notes. Use the modern measurements of the radial velocity with which these galaxies approach each other, $v_r\approx 109.3\pm 4.4$ km/s (Sohn et al. 2012) and distance between them, $r\approx 
770\pm 40$ kpc (van der Marel & Guhathakurta 2008) to estimate the sum of the MW and M31 masses.


### Solution



The Kahn-Woltjer method allows us to estimate the total (not only stellar mass) of the MW and M31 (Andromeda) galaxies, by assuming that the two galaxies are two point masses and solving the equations of their (relative) motion. 

There are assumtions we have to make in order to write down and solve the parametric equations of motion of the two galaxies. Assuming that their relative motion is $\mathbf{linear}$, (eccentricity $e=1$ or linear momentum $L=0$) the equations of motion can be written as:

\begin{equation}
r = \alpha (1 - \cos \eta), \qquad t = \sqrt{\frac{a^3}{GM}}(\eta - \sin\eta), \qquad \frac{dr}{dt} = \sqrt{\frac{GM}{\alpha}}\frac{\sin\eta}{1 - \cos\eta}
\end{equation}

where $r$ is the distance between the galaxies, $a$ the semi-major axis of the "orbit", $t$ the time since passed since the galaxies where at a distance $r=0$ (thus, this is almost equal to the age of the Universe) and $\eta$ is known as the eccentricity anomaly or development angle and parametrizes the orbit.

The equations can be combined in order to get the following equation:

\begin{equation}
\frac{t}{r}\frac{dr}{dt} = \frac{\sin \eta (\eta - \sin \eta)}{(1 - \cos \eta)^2}.
\end{equation}

Using the estimates of the current radial velocity ($dr/dt = v_r$), distance $r \cong 770$ kpc) and the time as the age of the Universe $t = t_U$, we can (numerically) solve the above algebraic equation in order to find the present value of the development angle $\eta$. Then we can go back, and using that value and the values for time and distance between the galaxies we can solve for the semimajor axis and the combined mass of the system of M31 + MW.

Using the stated values and the age of the universe: $t_U = 13.6$ Gyrs, we solve numerically the last equation:


```python
from scipy.optimize import fsolve
import numpy as np


def examplefunc(x, a):
    return np.sin(x)*(x - np.sin(x)) + a*(1.0 - np.cos(x))**2
    




#define time, distance and velocity. Be careful in units

#time in sec
#t = 365.0*24.0*3600.0*(13.6e9/4)
t = 365.0*24.0*3600.0*(13.6e9)
                                                      
#distance in km 
r = 770*(3.086e16)

# v in km/s 
v = 109.3

a = t*v/r
print(a)

x = fsolve(examplefunc, x0=4., args=(a))[0]
print "numerical solution is x=%.3f, f(x)=%.g"%(x,examplefunc(x, a))

a = r/(1.-np.cos(x))/3.086e16

print("a, r_apo = %.2f, %2.f (kpc)"%(a, 2.*a))
```

The first non-zero numerical solution is:

\begin{equation}
\eta = 3.597
\end{equation}

Using this value, we can solve for $\alpha$:

\begin{equation}
\alpha = \frac{r}{1 - \cos\eta} = \frac{770}{1 - \cos(3.597)} \,\mbox{kpc} = 405.67  \,\, \mbox{kpc}
\end{equation}


The parameter $\beta \equiv \sqrt{\frac{a^3}{G M}}$ can be calculated from:

\begin{equation}
\beta = \frac{t}{\eta - \sin \eta} = \frac{13.6}{3.597 - \sin(3.597)} = 3.368\,\, \mbox{Gyr}
\end{equation}

Solving for  $M = M_{MW} + M_{M31}$ we have:

\begin{equation}
M = \frac{\alpha^3}{G \beta^2}
\end{equation}

Plugging the numbers, we have (let's write a script to do correctly the unit conversion): 

```python
import numpy as np

#Use SI in order to do the calculation

G = 6.674e-11 

kpc = 3.086e19

alpha = (515.4)*kpc 

#year in sec

year = 365.0*24.0*3600

beta = 13.6e9*year/(x - np.sin(x))# (3.368e9)*year

#Calculate mass in kg

mass = np.power(alpha,3.0)/(G*np.power(beta,2.0))

#define solar mass

solmass = 1.99e30

#mass in solar masses

mass = mass/solmass

print(mass)
```

### Extra credit

The Hubble scale is defined as $c/H_0$ where $H_0\approx 70$ km/s/Mpc is the value of the Hubble "constant" at $z=0$. The mean density of matter in the universe can be expressed as $\bar{\rho}_{\rm m}=\Omega_{\rm m,0}\rho_{\rm cr,0}$, where current measurements indicate $\Omega_{\rm m,0}\approx 0.3$ and $\rho_{\rm cr,0}\equiv 3H_0^2/(8\pi G)$ is the critical density of the universe at $z=0$. Given these facts and your estimate for $M(250\ \mathrm{kpc})$ in the exercise 2 above, how many galaxies like the Milky Way could potentially form in the sphere of radius $c/H_0$?

### Solution

Let's calculate the total mass inside a sphere of  radius  $c/H_0$, assuming that it has a constant density equal to the mean density of the Universe. Then we can divide by the mass $M$(250 kpc) we found before, $M(250kpc)=1.25 \times10^{12} \,\, M_\odot$, to find the number of "Milky Ways" that can be form in that sphere.

Let's write a script in order to do the calculation.
 


```python
import numpy as np

#Calculation of rho, critical

#Let's calculate everything in SI

Gn = 6.674e-11

H0 = 70.0/(3.086e19)

omegami = 0.3

rhocrit = 3.0*(np.power(H0,2))/(8.0*(np.pi)*Gn)


#mean density

rmean = omegami*rhocrit

#Calculate the total mass of the sphere

c = 3.0e8

radius = c/H0 


totmass = rmean*(4.0*np.pi/3.0)*(radius**3.0)

#find in solar masses

totmass = totmass/(1.99e30)


#Mass of milky way, calculated above

MMW = 1.25e12

#find number of "Milky Ways"

Numb = totmass/MMW

print("The number of galaxies like the Milky Way could potentially form in the sphere of radius  c/H0 is:")
print(Numb)
```

Thus we predict a number of $\cong 1.07 \times 10^{10}$  galaxies like the Milky Way inside the Hubble radius.
