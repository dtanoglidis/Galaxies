### Solution 


###### a) 

At redshift of $z_i = 100$, we have to a very very good approximation  $\Omega_{\rm m}(z_i) =1$ (EdS Universe). Thus, any perturbation $\delta_i > 0$ is going to collapse ultimately. (If we want an exact value is $\delta_i >  \Omega_{\rm m}^{-1}(z_i)-1$  for cosmologies with  $\Omega_{\rm m}(z_i) <1$.

Let's calculate the range of values for which the perturbation collapses before $z=0$. We will consider both $\Omega_{\rm m,0} = 1$ and $\Omega_{\rm m,0} = 0.3$ cosmologies (both with $\Omega_{\rm \Lambda, 0} = 0$ and the same value of the Hubble constant).


Since the overdensity collapses today ($z=0$), the linearly extrapolated -- to the present day -- overdensity should be equal to the critical linear overdenisty for collapse. Denoting the linearly extrapolated overdensities with a tilde: $\tilde{\delta}$, we thus have:

\begin{equation}
\tilde{\delta}(z=0) = \tilde{\delta}_{coll} \cong 1.686 \qquad (1)
\end{equation}

This value is almost independent on the value of $\Omega_{\rm m}$, so we will use it in both cases.

Now, the overdensity $\delta_i$ at $z=100$ that by $z=0$ will have the --linealy extrapolated-- value $\tilde{\delta}_{coll}$ (and thus would collapse by then) can be calculated using the $\mathbf{growth\,\, factor}$ as:

\begin{equation}
\delta_i = \tilde{\delta}_{coll} D_{+}(z_i),  
\end{equation}

where the normalization of the growth factor is such that $D_{+}(z = 0) = 1$. The growth factor can be calculated for every cosmological model. 

We will the colossus routine to get the growth factor at $z=100$ for the two cosmologies of interest.



First for the $\Omega_{\rm m} = 1$ cosmology:

```python
import numpy as np
# import Benedikt Diemer's colossus package
from colossus.cosmology import cosmology
# set cosmology parameters
params1 = {'flat': False, 'H0': 70.0, 'Om0': 1.0, 'relspecies': False, 'OL0': 0, 
          'Ob0': 0.049, 'sigma8': 0.8, 'ns': 0.968, 'print_warnings': False}
cosmo1 = cosmology.setCosmology('mycosmo', params1 )
params2 = {'flat': False, 'H0': 70.0, 'Om0': 0.3, 'relspecies': False, 'OL0': 0, 
          'Ob0': 0.049, 'sigma8': 0.8, 'ns': 0.968, 'print_warnings': False}
cosmo2 = cosmology.setCosmology('mycosmo', params2 )





z = 100.0
# growth factor of the universe normalize to 1.0 at z=0
Dz = cosmo1.growthFactor(z)

print("z = %.3f; D_+(z) = %.3e"%(z, Dz))
```

For the $\Omega_{\rm m} = 1$ cosmology, we thus have:

\begin{equation}
D_{+}(z=100) = 9.899 \times 10^{-3}
\end{equation}

Thus, from equation (1) we have that:

\begin{equation}
\delta_i \cong 0.0167 \qquad (2)
\end{equation}

Let's calculate the growth factor and $\delta_i$ in the case of $\Omega_{\rm m} = 0.3$ cosmology now:

```python
z = 100.0
# growth factor of the universe normalize to 1.0 at z=0
Dz = cosmo2.growthFactor(z)

print("z = %.3f; D_+(z) = %.3e"%(z, Dz))
```

For the $\Omega_{\rm m} = 0.3$ cosmology, we thus have:

\begin{equation}
D_{+}(z=100) = 2.167 \times 10^{-2}
\end{equation}

Thus, from equation (1) we have that:

\begin{equation}
\delta_i \cong 0.0365 \qquad (3)
\end{equation}

Higher than that in the previous case, as expected.

Note that we use $h = 0.7$ throughout, so we can easily get the mass in sollar masses by using this value for that factor.


In order to calculate how rare are perturbations of such mass at $z_i = 100$, we will use that the overdensity field, smoothed at mass scale $M$ at $z = 100$ follows a gaussian probability density distribution (PDF):

$$
P(\delta_{M})=\frac{1}{\sqrt{2\pi}\sigma(M)}\exp\left[-\frac{\delta^2_{M}}{2\sigma^2(M)}\right].
$$

$\sigma(M)$ is the variance of the smoothed field (and is equivalent to $\sigma(R)$ for the relevant length\radius scalse, since there is a 1 to 1 correspondence between radius and mass.

In order to estimate how rare is the perturbation, we can calculate the ratio of the probability to find at that epoch amplitude $\delta_i$ at the smoothed overdensity field, to find zero overdensity ($\delta = 0$), thus:


\begin{equation}
\frac{P(\delta_i)}{P(0)} =  \exp \left[ -\frac{\delta_i^2}{2\sigma^2(M)} \right].
\end{equation}


Let's compute again this for the two cases -- $\Omega_{\rm m,0} = 1.0$ and $\Omega_{\rm m,0} = 0.3$ cosmologies.

As usual, start with the $\Omega_{\rm m,0} = 1.0$ cosmology, for which we have $\delta_i = 0.0167$.



```python

delti = 0.0167

R = 1.5 # comoving in h^{-1} Mpc
z = 100.0
# compute rms fluctuations on comoving scale R at redshift z
sig = cosmo1.sigma(R, z=z, j=0) 




factor = np.exp( - (delti / sig)**2.0/2.0)

                
print("Relative probability for the perturabation of such mass scale in a Omega_m = 1.0 cosmology is:")
print(factor)

```

Let's now calculate the same thing for the case of  an $\Omega_{\rm m,0} = 0.3$ cosmology, where $\delta_i = 0.0365$

```python
delti = 0.0365

R = 1.5 # comoving in h^{-1} Mpc
z = 100.0
# compute rms fluctuations on comoving scale R at redshift z
sig = cosmo2.sigma(R, z=z, j=0) 




factor = np.exp( - (delti / sig)**2.0/2.0)

                
print("Relative probability for the perturabation of such mass scale in a Omega_m = 0.3 cosmology is:")
print(factor)
```

###### c)

Initial velocity is given (see notes) by:

\begin{equation}
\displaystyle
v_i \equiv \dot{r}_i  = H(t_i)r_i - \frac{r_i}{3}\frac{\delta_i}{1 + \delta_i} \frac{\dot{D}_+(t_i)}{D_{+}(t_i)} \qquad (6)
\end{equation}

where $H(t_i)$ the Hubble parameter at the initial time, $r_i = a(t_i)R_i$, $\delta_i$ the initial overdensity  and $D_{+}(t_i)$ the growth factor.

Note that at zeroth order, the velocity is just the Hubble expansion rate.

We will calculate the initial velocity in km/s. Now, we will use, where needed the value of $h = 0.7$ in order to our final results not to have any $h$ ambiguity.


Again, we have to calculate this for both cosmologies, $\Omega_{\rm m,0} = 1.0$ and $\Omega_{\rm m,0} = 0.3$.


We start - what a surprise! with the case of $\Omega_{\rm m,0} = 1.0$

```python
delti = 0.0167
h = 0.7

R = 1.5 # comoving in h^{-1} Mpc
z = 100.0

# compute expansion rate H(z) at redshift z in km/s/Mpc
Hz = cosmo1.Hz(z) 

# age of the universe at z
tz = cosmo1.age(z)
# growth factor of the universe normalize to 1.0 at z=0
Dz = cosmo1.growthFactor(z)
# dDdt = dD_+/dt = dD_+/dz*dz/dt in 1/Gyr
dDdt = cosmo1.growthFactor(z, derivative=1) * cosmo1.age(tz, derivative=1, inverse=True)
# DDr = dD_+(ti)/dt/ D_+(ti) in 1/Gyr
DDr = dDdt/cosmo1.growthFactor(z)


#Calculate r_i in Mpc 
rinit = R/(h*(1.0 + z))

#Applying equation 6 as it is, in the second part we will have units of Mpc/Gyr. We have to include a 
#conversion factor in order to get km/s

#First Mpc to km then Gyr to s

convfactor = (3.086e19)/(3.15e16)

#Now, we are ready to apply 6

v = Hz*rinit - (rinit/3.0)*(delti/(1.0 + delti))*DDr*convfactor

print("Initial velocity in km/s is:")

print(v)
```

The initial velocity at the outermost shells of the perturbation, in a $\Omega_{\rm m,0}= 1.0$ cosmology is thus

\begin{equation}
v_i = 1499.168 \, \, \mbox{km/s} \qquad (7)
\end{equation}

```python
delti = 0.0365
h = 0.7

R = 1.5 # comoving in h^{-1} Mpc
z = 100.0

# compute expansion rate H(z) at redshift z in km/s/Mpc
Hz = cosmo2.Hz(z) 

# age of the universe at z
tz = cosmo2.age(z)
# growth factor of the universe normalize to 1.0 at z=0
Dz = cosmo2.growthFactor(z)
# dDdt = dD_+/dt = dD_+/dz*dz/dt in 1/Gyr
dDdt = cosmo2.growthFactor(z, derivative=1) * cosmo1.age(tz, derivative=1, inverse=True)
# DDr = dD_+(ti)/dt/ D_+(ti) in 1/Gyr
DDr = dDdt/cosmo2.growthFactor(z)


#Calculate r_i in Mpc 
rinit = R/(h*(1.0 + z))

#Applying equation 6 as it is, in the second part we will have units of Mpc/Gyr. We have to include a 
#conversion factor in order to get km/s

#First Mpc to km then Gyr to s

convfactor = (3.086e19)/(3.15e16)

#Now, we are ready to apply 6

v = Hz*rinit - (rinit/3.0)*(delti/(1.0 + delti))*DDr*convfactor 

print("Initial velocity in km/s is:")

print(v)
```

###### d) Plot of R(t) and a(t)

In a universe without a cosmological constant ($\Lambda = 0$) the equation of motion for an overdensity is:

\begin{equation}
\ddot{r} = - \frac{GM}{r} \qquad (9)
\end{equation}

This has the parametric solution:

\begin{equation}
\displaystyle
r = \frac{r_{ta}}{2}\left(1 - \cos \eta \right), \quad t = \frac{t_{ta}}{\pi}\left(\eta - \sin \eta \right) \qquad (10)
\end{equation}

where $t_{ta}$ and $r_{ta}$ the time and radius at $\mathbf{turnaround}$ (when the perturbation reaches the maximum size), respectively. $\eta$ is a development angle which runs from $0$ to $2 \pi$.

These are connected to the initial conditions $r_i$ and $v_i$.  First, using these we can calculate the specific energy (energy per unit mass):

\begin{equation}
E = \frac{1}{2} v_i^2 - \frac{GM}{r_i} \qquad (11)
\end{equation}

Then the turnaround radius can be calculated:

\begin{equation}
r_{ta} = \frac{GM}{| E |} \qquad (12)
\end{equation}

And then we can use this to calculate the turnaround time:

\begin{equation}
t_{ta} = \left(\frac{\pi^2 r_{ta}^3}{8 G M} \right)^{1/2} \qquad (13)
\end{equation}


What we have to do is actually for a given $R_i$ and $\delta_i$ to repeat the steps above - calculate mass, $r_i$, $v_i$ and use them in order to find $R(t)$ using the parametric solution. 

We will also plot $R_i a(t)$ for comparison.


Let's start -- again! -- with $\Omega_{\rm m,0} = 1.0$ cosmology.  Use $\delta_i = 0.025$ -- higher than the overdensity for collapse today. Also, $R_i = 1.5 h^{-1}$ Mpc.

```python
R = 1.5 # comoving in h^{-1} Mpc
z = 100.0


M = cosmo1.lagrangianM(R) # returns mass in /h Msun 

print("Mass in M_sol h^{-1}")
print(M)
```

I will now argue that I am $\mathbf{not}$ going to use that mass when using (11-12-13) in order to calculate $r_{ta}$ and $t_{ta}$. 

The reason is that these expressions are $\mathbf{very}$ sensitive in the value of each of the parameters we use, so we have to be very careful. To get the mass, I will use an expression consistent with our definition of inital radius and  the overdensity. Mass is given by: 

\begin{equation}
M = \frac{4 \pi}{3}  r_i^3 \rho_i = \frac{4 \pi}{3} r_i^3 \overline{\rho}_i (1 + \delta_i) = \frac{4 \pi}{3}R_i^3 \overline{\rho}_{i,0}(1 + \delta_i) \qquad (14) 
\end{equation}

We'll use (14) to calculate the mass of the perturbation.

```python
delti = 0.025 

#comoving radius in kpc/h
R = 1.5*1000

z = 0.0
# density of matter at redshift z
rhoz = cosmo1.rho_m(z) # in h^2 Msun/kpc^3


M = (4.0*np.pi/3.0)*(R**3.0)*rhoz*(1.0 + delti)

print("Mass in M_sol h^{-1}")
print(M)
```

We see that the mass we have now:

\begin{equation}

\end{equation}

is different than the lagrangian mass we calculated above. Although not much different, this difference is very important when calculating the values of $t_{ta}$ and $r_{ra}$. 

We have to calculate $r_i$ and $v_i$ in Mpc and km/s, respectively. Working as above, we calculate them:

```python
delti = 0.025
h = 0.7

R = 1.5 # comoving in h^{-1} Mpc
z = 100.0

# compute expansion rate H(z) at redshift z in km/s/Mpc
Hz = cosmo1.Hz(z) 

# age of the universe at z
tz = cosmo1.age(z)
# growth factor of the universe normalize to 1.0 at z=0
Dz = cosmo1.growthFactor(z)
# dDdt = dD_+/dt = dD_+/dz*dz/dt in 1/Gyr
dDdt = cosmo1.growthFactor(z, derivative=1) * cosmo1.age(tz, derivative=1, inverse=True)
# DDr = dD_+(ti)/dt/ D_+(ti) in 1/Gyr
DDr = dDdt/Dz


#Calculate r_i in Mpc 
rinit = R/(h*(1.0 + z))

#Applying equation 6 as it is, in the second part we will have units of Mpc/Gyr. We have to include a 
#conversion factor in order to get km/s

#First Mpc to km then Gyr to s

convfactor = (3.086e19)/(3.15e16)

#Now, we are ready to apply 6

v = Hz*rinit  -  (rinit/3.0)*(delti/(1.0 + delti))*DDr*convfactor

print("initial radius in Mpc is")
print(rinit)


print("Initial velocity in km/s is:")

print(v)

```

###### e)  

Here we will calculate the density contrast:

\begin{equation}
\delta(t) \equiv \frac{{\rho}(t)}{\overline{\rho}_m(t)}-1, \qquad (15)
\end{equation}

in the two different cosmological models, and we will compare to the linear theory expression:

\begin{equation}
\delta(t) = \delta_i \frac{D_+(t)}{D_+(t_i)} \qquad (16)
\end{equation}

In the above, we will need to calculate $\rho(t)$, which is given by:

From virial theorem $\Rightarrow$ perturbation collapses not to a singularity but to a bound structure obeying the virial theorem:

\begin{equation}
2K + W = 0, \qquad (18)
\end{equation}

Remember that total energy: $E = W + K$ so, finally: $E_{coll} =  \frac{1}{2}W_{coll}$. 
At turnaround $K = 0$, so $E_{ta} = W_{ta}$. But energy is conserved, so $E_{ta} = E_{coll}$, which gives:

\begin{equation}
W_{coll} = 2W_{ta} \Rightarrow -\frac{GM}{r_{coll}} = - \frac{GM}{r_{ta}} \Rightarrow r_{coll} = \frac{1}{2}r_{ta}
\end{equation}

Since the mass $M$ is constant, we have that: $\rho_{coll} = 8 \rho_{ta}$.

So, we can easilly calculate $\rho_{ta}$ and through it, $\rho_{coll}$.

We can also calculate the density of the background Universe at the time of collapse; $t_{coll} = 2t_{ta}$, and through this the overdensity. 

$\mathbf{note}$: One may argue that this is not the most general way to do the calculation; here for we will calculate this for the two specific cases we have worked out so far. However, the results hold generally. So, let's calculate the collapse overdensity for the two mentioned cosmologies. 


```python
#Start for $\Omega_mi =1 cosmology

h = 0.7

#Mass in units h^-1 M_sol
Mass1 = 4.02092016476e12
delti1 = 0.025

#turnaround radius in Mpc
rta1 = 0.5269266584675234 
#turnaround time in Gyr 
tta1 =  2.6444668765936585

#calculation of collapse density

rhocoll1 = 8.0*(Mass1/h)/((4.0*np.pi/3)*(rta1**3.0))

z1 = cosmo1.age(2*tta1, derivative = 0, inverse = True)

rhoback1 = cosmo1.rho_m(z1) # in h^2 Msun/kpc^3

#convert to the same units as rho_coll

rhoback1 = (h**2.0)*rhoback1*(10**9.0)

deltacoll1 = rhocoll1/rhoback1 - 1.0



print("Overdensity for collapse in an Omega_m = 1.0 universe:")
print(deltacoll1)



# Then for $\Omega_mi = 0.3 cosmology


#Mass in units h^-1 M_sol
Mass2 = 1.22863628839e+12
delti2 = 0.044

#turnaround radius in Mpc
rta2 = 0.44652226336477174
#turnaround time in Gyr 
tta2 = 3.7318880493254865

#calculation of collapse density

rhocoll2 = 8.0*(Mass2/h)/((4.0*np.pi/3)*(rta2**3.0))

z2 = cosmo2.age(2*tta2, derivative = 0, inverse = True)

rhoback2 = cosmo2.rho_m(z2) # in h^2 Msun/kpc^3

#convert to the same units as rho_coll

rhoback2 = (h**2.0)*rhoback2*(10**9.0)

deltacoll2 = rhocoll2/rhoback2 - 1.0


print("Overdensity for collapse in an Omega_m = 0.3 Universe:")
print(deltacoll2)



```

Thus, the overdensity for collapse in $\Omega_{\rm m,0} = 1.0$ cosmology is:

\begin{equation}
\delta_{coll} \cong 177.1 
\end{equation}

and in $\Omega_{\rm m,0} = 0.3$ cosmology is:

\begin{equation}
\delta_{coll} \cong 341.8
\end{equation}

We see that in the low-density cosmology, the overdensity required for collapse is significantly higher than the $\Omega_{\rm m,0} = 0.3$ cosmology. This is reasonable, since the perturbation is more difficult to collapse in a more diffuse environment and a higher density contrast is required.

###### a) 

In the first part of of this problem I will use the abundance matching ansatz, as described above,  to derive the $M_* - M_h$ relation, today ($z = 0$). 

The steps are the following:

$\bullet$ First for a range of masses, tabulate and spline the MST cumulative mass function.

$\bullet$ Repeat the same for stellar mass function as calculated in HW2

$\bullet$ Use the abundance matching ansatz, by finding for each value of $M_h$ the value of $M_*$ for which $n_g(>M_*) = n_h(>M_h)$.

```python
from colossus.cosmology import cosmology
import matplotlib.cm as cm
import matplotlib.colors as colors
from scipy.interpolate import UnivariateSpline
from scipy.integrate import quadrature


# First: create/tabulate MST halo mass function

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology	
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
#cosmo.interpolation = False


def fsub(mh, z):
    """
    subhalo fraction calibration from the Appendix G of Behroozi et al. 2013
    as a function of Mpeak - eqs G9-10
    input: mh = halo mass in Msun
           z  = redshift
    """
    aexp = 1.0/(1.0+z)
    aexp2 = aexp*aexp; aexp3 = aexp2*aexp; aexp4 = aexp3*aexp
    Ca = -1.91 + 6.23*aexp - 15.07*aexp2 + 15.02*aexp3  - 5.29*aexp4
    Mcutoff = 10.66 + 15.93*aexp - 21.39*aexp2 + 18.20*aexp3 - 8.21*aexp4
    return 10.**Ca*np.log10(10.**Mcutoff/mh)

#Value of h
hub = 0.7

z0 = 0.0
# Here we use just one redshift
zd = 0.0

# grid of halo masses in /h Msun
lmh = np.arange(6.,15.5,0.05)
mh = np.power(10.,lmh)
# compute Lagrangian radius corresponding to perturbation of mass mh in /h Mpc
rh = cosmo.lagrangianR(mh)
# compute rms fluctuation sigma(R) using default Eisenstein & Hu 1998 power spectrum and tophat filter
sig = cosmo.sigma(rh,z0)
# and its logarithmic derivative dlog(sigma)/dlogR
dsdr = cosmo.sigma(rh, z0, derivative='True')
# dln(sigma)/dlnR -> dln(sigma)/dlnM, M = (4*pi/3) rho_mean * R^3
dsdm = dsdr / 3.0



#  SMT01 mass function parameters
#
q_smt = 0.3; A_smt = 0.322; a_smt = 0.707

nu = cosmo.peakHeight(mh, zd)
nup = np.sqrt(a_smt)*nu
dsdmz = dsdm*cosmo.growthFactor(zd)
nsmt2 = A_smt*np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh*(1.0+1./nup**(2.*q_smt))*nup*np.abs(dsdm)*np.exp(-0.5*nup**2)


#We have created the halo mass function. We will multiply it by 1+f_sub, the fraction of subhalos

#Also, in order to go from dn/dlnM to dn/dlog_10 M --which is the most convenient thing to do--  we multyply by ln10


halomassfnc = nsmt2*(1 + fsub(mh,zd))*np.log(10.0)

#Let's now  spline the function under the integral itself, so we can use integral method of the spline 
halomfncspline = UnivariateSpline(lmh, halomassfnc, s=0.0)


#calculation of the cumulative halo mass function

halocum = np.zeros_like(lmh)

for i, lmhd in enumerate(lmh):
    halocum[i] = halomfncspline.integral(lmhd,np.inf)

    
#Now, it is time to do the same thing for the stellar mass function.
```

For the stellar mass function, we will use the best Schechter fit, as calculated in PS2, with the parameters used there. Although it is not 100$\%$ accurate, especially in higher masses, it gives us the opportunity to have a neat solution, without the need to import and read the full catalogue of galaxies and calculate the luminosity function and mass function from it.

```python

#grid of stellar masses in 1/h M_sun 
#dividing by h gives units 1/h since initially we have units 1/h^2 - SEE PS2
h = 0.7
mstel = np.arange(7.0,12.0,0.05)


popt = np.zeros(3)
popt[0] = 4.00e-03
popt[1] = 10.99
popt[2] = -0.28

from scipy.optimize import curve_fit

def lf_schechter(L, phis, Lstar, alpha):
    mus = np.power(10,L)/np.power(10,Lstar)
    #print "phis, Lstar, alpha", phis, Lstar, alpha
    return np.log(10.)*phis*mus**(alpha)*np.exp(-mus)

Lsexp = int(popt[1]); Lsnum = popt[1] - Lsexp

#calculation of stellar mass function 

stellarmassfnc = lf_schechter(mstel, popt[0], popt[1], popt[2])
```

We have now the stellar mass function, $dn_{stell}/dM_*$; we want to repeat the same steps as above. We will spline $dn/dM$ as a function on $M$.

```python
stellarmass =  mstel

stellarfncspline = UnivariateSpline(stellarmass, stellarmassfnc, s=0.0)

#calculation of the cumulative stellar mass function

stellarcum = np.zeros_like(stellarmass)

for i, lmhd in enumerate(stellarmass):
    stellarcum[i] = stellarfncspline.integral(lmhd,np.inf)
    
```

We are now ready to match the stellar and halo mass functions using abundance matching.

```python
stellarcum[stellarcum<=0] = 1.e-9

# note that the first variable in the spline must be increasing, so we need to invert zcum
stellarcumr = stellarcum[::-1]; stellarmassr = stellarmass[::-1]


matchspl = UnivariateSpline(stellarcumr, stellarmassr, s=0.0)

stelmassmatch = matchspl(halocum)

fig = plt.figure(figsize = (5.5,4.5))
plt.plot(np.power(10,lmh), np.power(10,stelmassmatch), c='c', linewidth ='2', label=r'Stelar mass VS Halo mass, $z=0$')
plt.xscale('log'); plt.yscale('log')
plt.ylim(1.0e6,1.0e12)
plt.xlim(1.0e10,1.0e15)

plt.legend(frameon=False, loc='lower right', fontsize=9)
plt.xlabel('Halo mass $M_h \,\,\, [ h^{-1} M_\odot ]$')
plt.ylabel('Stellar mass  $M_* \,\,\, [h^{-2} M_\odot]$')

```

```python
import numpy as np
from matplotlib import pyplot as plt
import math
from scipy.interpolate import UnivariateSpline
import matplotlib.cm as cm
import matplotlib.colors as colors


def nmstz(ms, z):
    """
    compute analytic double Schechter representation of the observed stellar mass function as a function
    of redshift using calibration of Leja et al. 2015, 798, 115 (see eqs in their S 2)
    note that this calibration is for 3<z<0 - I checked it gives grossly incorrect results at z>3
    
    input: ms - array of stellar masses
            z - redshift
    output: dn/dlog10(M*) in dex^-1 Mpc^-3
    """
    if z <= 3.0:
        z2 = z * z; lms = np.log10(ms)
        lphi1 = -2.46 + 0.07*z - 0.28*z2 
        lphi2 = -3.11 - 0.18*z - 0.03*z2
        lmst  = 10.72 - 0.13*z + 0.11*z2
        mst = np.power(10.,lmst); mus = ms/mst
        a1 = -0.39; a2 = -1.53
        nmstz = np.log(10.)*np.exp(-mus)*(mus)*(10.**lphi1*mus**a1+10.**lphi2*mus**a2)
    else:
        """
        high-z SMF estimates from CANDELS and S-CANDELS from the talk by Song 
        http://www.astrodeep.eu/wp-content/uploads/2015/02/MSong.sesto_.Jan2015.pdf
        """
        z_h     = [4.0, 5.0, 6.0, 7.0]
        lmst_z  = [10.68, 10.58, 10.22, 10.45]
        ah_z    = [-1.59, -1.68, -1.89, -1.93]
        lphi_z = [-3.77, -3.99, -4.34, -4.94]

        lmst  = UnivariateSpline(z_h, lmst_z)
        ahz   = UnivariateSpline(z_h, ah_z)
        lphiz = UnivariateSpline(z_h, lphi_z)
        a = ahz(z); mst = np.power(10., lmst(z)); phiz = np.power(10.,lphiz(z))
        mus = ms/mst
        nmstz = np.log(10.)*phiz*np.exp(-mus)*(mus)**(a + 1.0)

    return nmstz


lms = np.arange(7., 13.0, 0.01)
ms = np.power(10., lms)

philejaz00 = nmstz(ms, 0.0)
philejaz01 = nmstz(ms, 0.1)
philejaz05 = nmstz(ms, 0.5)
philejaz10 = nmstz(ms, 1.0)
philejaz15 = nmstz(ms, 1.5)
philejaz20 = nmstz(ms, 2.0)
philejaz25 = nmstz(ms, 2.5)
candelsz35 = nmstz(ms, 3.5)
candelsz50 = nmstz(ms, 5.0)
candelsz60 = nmstz(ms, 6.0)
candelsz70 = nmstz(ms, 7.0)
candelsz80 = nmstz(ms, 8.0)


plt.plot(ms, philejaz00, c ='crimson', label=r'$z=0.0$' )
plt.plot(ms, philejaz01, c='lime', label=r'$z=0.1$')
plt.plot(ms, philejaz05, c = 'teal', ls='--', label=r'$z=0.5$')
plt.plot(ms, philejaz10, c = 'purple', ls = '-.', label =r'$z=1.0$')
plt.plot(ms, philejaz15, c = 'maroon', label = r'$z=1.5$')
plt.plot(ms, philejaz20, c = 'navy', ls ='--', label=r'$z=2.0$')
plt.plot(ms, philejaz25, c = 'yellow', label=r'$z=2.5$')
plt.plot(ms, candelsz35, c = 'aqua', label= r'$z=3.5$')
plt.plot(ms, candelsz50, c = 'violet', ls = '--', label = r'$z=5.0$')
plt.plot(ms, candelsz60, c = 'olive', label = r'$z=6.0$')
plt.plot(ms, candelsz70, c = 'gray', label = r'$z =7.0$')
plt.plot(ms, candelsz80, c = 'indigo', label = r'$z=8.0$')

plt.ylim(1.0e-9,1.0e-1)
plt.xlim(1.0e8,5.0e12)
plt.legend(frameon=False, loc='lower left', fontsize=8)
plt.xscale('log'); plt.yscale('log')

plt.xlabel('Stellar mass  M_* [$M_\odot$]')
plt.ylabel('Number density [Mpc$^{-3}$ dex$^{-1}$]')

plt.show()
```

```python
# Calculation of Halo mass function for a range of redshifts. Choose 0.0, 1.0, 2.0 , 3.0, 5.0, 7.0
# We can produce more, but we chose these characteristic examples

#Let's create mass function for these cases 

#It will be the total halo mass function, multiplied by the subhalo fraction 1+f and ln10


from colossus.cosmology import cosmology
import matplotlib.cm as cm
import matplotlib.colors as colors
from scipy.interpolate import UnivariateSpline
from scipy.integrate import quadrature


# First: create/tabulate MST halo mass function

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology	
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
#cosmo.interpolation = False


def fsub(mh, z):
    """
    subhalo fraction calibration from the Appendix G of Behroozi et al. 2013
    as a function of Mpeak - eqs G9-10
    input: mh = halo mass in Msun
           z  = redshift
    """
    aexp = 1.0/(1.0+z)
    aexp2 = aexp*aexp; aexp3 = aexp2*aexp; aexp4 = aexp3*aexp
    Ca = -1.91 + 6.23*aexp - 15.07*aexp2 + 15.02*aexp3  - 5.29*aexp4
    Mcutoff = 10.66 + 15.93*aexp - 21.39*aexp2 + 18.20*aexp3 - 8.21*aexp4
    return 10.**Ca*np.log10(10.**Mcutoff/mh)

#Value of h
hub = 0.7

z0 = 0.0

# grid of halo masses in /h Msun
lmh = np.arange(6.,15.5,0.05)
mh = np.power(10.,lmh)
# compute Lagrangian radius corresponding to perturbation of mass mh in /h Mpc
rh = cosmo.lagrangianR(mh)
# compute rms fluctuation sigma(R) using default Eisenstein & Hu 1998 power spectrum and tophat filter
sig = cosmo.sigma(rh,z0)
# and its logarithmic derivative dlog(sigma)/dlogR
dsdr = cosmo.sigma(rh, z0, derivative='True')
# dln(sigma)/dlnR -> dln(sigma)/dlnM, M = (4*pi/3) rho_mean * R^3
dsdm = dsdr / 3.0



#  SMT01 mass function parameters
#
q_smt = 0.3; A_smt = 0.322; a_smt = 0.707

def NSMTmassfnc(mh, zd):
    nu = cosmo.peakHeight(mh, zd)
    nup = np.sqrt(a_smt)*nu
    dsdmz = dsdm*cosmo.growthFactor(zd)
    nsmt2 = A_smt*np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh*(1.0+1./nup**(2.*q_smt))*nup*np.abs(dsdm)*np.exp(-0.5*nup**2)
    halomassfnc = nsmt2*(1 + fsub(mh,zd))*np.log(10.0)
    return halomassfnc

#Let's use to calculate mass functions in the above mentioned redshifts

halomassz0 = NSMTmassfnc(mh, 0.0)
halomassz1 = NSMTmassfnc(mh, 1.0)
halomassz2 = NSMTmassfnc(mh, 2.0)
halomassz3 = NSMTmassfnc(mh, 3.0)
halomassz5 = NSMTmassfnc(mh, 5.0)
halomassz7 = NSMTmassfnc(mh, 7.0)



#Let's now  spline the halo mass function under the integral itself, so we can use integral method of the spline 
halosplinez0 = UnivariateSpline(lmh, halomassz0, s=0.0)
halosplinez1 = UnivariateSpline(lmh, halomassz1, s=0.0)
halosplinez2 = UnivariateSpline(lmh, halomassz2, s=0.0)
halosplinez3 = UnivariateSpline(lmh, halomassz3, s=0.0)
halosplinez5 = UnivariateSpline(lmh, halomassz5, s=0.0)
halosplinez7 = UnivariateSpline(lmh, halomassz7, s=0.0)

#calculation of the cumulative halo mass functions

halocumz0 = np.zeros_like(lmh)
halocumz1 = np.zeros_like(lmh)
halocumz2 = np.zeros_like(lmh)
halocumz3 = np.zeros_like(lmh)
halocumz5 = np.zeros_like(lmh)
halocumz7 = np.zeros_like(lmh)

for i, lmhd in enumerate(lmh):
    halocumz0[i] = halosplinez0.integral(lmhd,np.inf)
    halocumz1[i] = halosplinez1.integral(lmhd,np.inf)
    halocumz2[i] = halosplinez2.integral(lmhd,np.inf)
    halocumz3[i] = halosplinez3.integral(lmhd,np.inf)
    halocumz5[i] = halosplinez5.integral(lmhd,np.inf)
    halocumz7[i] = halosplinez7.integral(lmhd,np.inf)
    
    

    
```

```python
#Although we have most of them, let's get here again the stellar mass functions for the same redshifts

lms = np.arange(7., 13.0, 0.01)
ms = np.power(10., lms)

stellarmass0 = nmstz(ms, 0.0)
stellarmass1 = nmstz(ms, 1.0)
stellarmass2 = nmstz(ms, 2.0)
stellarmass3 = nmstz(ms, 3.0)
stellarmass5 = nmstz(ms, 5.0)
stellarmass7 = nmstz(ms, 7.0)

#Let's now  spline the stellar mass function under the integral itself, so we can use integral method of the spline 

stellarsplinez0 = UnivariateSpline(lms, stellarmass0, s=0.0)
stellarsplinez1 = UnivariateSpline(lms, stellarmass1, s=0.0)
stellarsplinez2 = UnivariateSpline(lms, stellarmass2, s=0.0)
stellarsplinez3 = UnivariateSpline(lms, stellarmass3, s=0.0)
stellarsplinez5 = UnivariateSpline(lms, stellarmass5, s=0.0)
stellarsplinez7 = UnivariateSpline(lms, stellarmass7, s=0.0)


#calculation of the cumulative stellar mass functions
stellarcumz0 = np.zeros_like(lms)
stellarcumz1 = np.zeros_like(lms)
stellarcumz2 = np.zeros_like(lms)
stellarcumz3 = np.zeros_like(lms)
stellarcumz5 = np.zeros_like(lms)
stellarcumz7 = np.zeros_like(lms)


for i, lmsd in enumerate(lms):
    stellarcumz0[i] = stellarsplinez0.integral(lmsd,np.inf)
    stellarcumz1[i] = stellarsplinez1.integral(lmsd,np.inf)
    stellarcumz2[i] = stellarsplinez2.integral(lmsd,np.inf)
    stellarcumz3[i] = stellarsplinez3.integral(lmsd,np.inf)
    stellarcumz5[i] = stellarsplinez5.integral(lmsd,np.inf)
    stellarcumz7[i] = stellarsplinez7.integral(lmsd,np.inf)
    
    

stellarcumz0[stellarcumz0<=0] = 1.e-9
stellarcumz1[stellarcumz1<=0] = 1.e-9
stellarcumz2[stellarcumz2<=0] = 1.e-9
stellarcumz3[stellarcumz3<=0] = 1.e-9
stellarcumz5[stellarcumz5<=0] = 1.e-9
stellarcumz7[stellarcumz7<=0] = 1.e-9


# note that the first variable in the spline must be increasing, so we need to invert zcum
lmsr = lms[::-1]
stellarcumrz0 = stellarcumz0[::-1] 
stellarcumrz1 = stellarcumz1[::-1]
stellarcumrz2 = stellarcumz2[::-1]
stellarcumrz3 = stellarcumz3[::-1]
stellarcumrz5 = stellarcumz7[::-1]
stellarcumrz7 = stellarcumz7[::-1]



matchsplz0 = UnivariateSpline(stellarcumrz0, lmsr, s=0.0)
matchsplz1 = UnivariateSpline(stellarcumrz1, lmsr, s=0.0)
matchsplz2 = UnivariateSpline(stellarcumrz2, lmsr, s=0.0)
matchsplz3 = UnivariateSpline(stellarcumrz3, lmsr, s=0.0)
matchsplz5 = UnivariateSpline(stellarcumrz5, lmsr, s=0.0)
matchsplz7 = UnivariateSpline(stellarcumrz7, lmsr, s=0.0)


stelmassmatchz0 = matchsplz0(halocumz0)
stelmassmatchz1 = matchsplz1(halocumz1)
stelmassmatchz2 = matchsplz2(halocumz2)
stelmassmatchz3 = matchsplz3(halocumz3)
stelmassmatchz5 = matchsplz5(halocumz5)
stelmassmatchz7 = matchsplz7(halocumz7)

#Now, let's plot!!!!!
```

### Problem 3 - Galaxy evolution model

To get appreciation for processes that shape galaxy properties, we will build a fairly simple galaxy formation model. Many elements will be based on the model described in S 2.1-2.2 of Feldmann (2013) (some details can also be found in S 2 of Krumholz & Dekel 2012). It is essentially a system of ODEs with specified sink and source terms specified by the model. The ODEs can be integrated using one of the python integrators. The code used below uses scipy.integrate.odeint

We will start with the bare-bones model framework that I provide in the python code galaxy_model.py in code/ subdirectory. The code uses object-oriented features of python (namely class and methods), which are convenient tools for this purpose. If you are not familiar with classes, you can read some intro material here, for example (we will only use the most basic features and you can see everything that's needed in my code, except for inheritance, which we can discuss later).
Specifically, in galaxy_model.py a bare-bones galaxy model is implemented as a python class, which is convenient in this case because it allows to collect all the relevant quantities and functions and encapsulate them in a single object. This then makes it easy to create collections of model galaxies with a given set of parameters, or to initialize a model at any instance in time given the key parameters and use class methods to compute quantities of interest (like SFR) easily without the need to store them.

Below you can see example of how model galaxies are instantiated and used.

The model implements realistic evolution of halo mass, but simply assumes that baryons with universal fraction accrete onto galaxy instantly and then get converted into stars on a constant 1 Gyr time scale.

```python
import numpy as np
import code.galaxy_model as gm
from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
```

```python
from scipy.integrate import odeint

def test_galaxy_evolution(Minit, sfrmodel, cosmo, verbose):
    zg = np.linspace(20., 0., 40) # grid of redshifts
    t_output = cosmo.age(zg) # and corresponding times

    # instantiate a galaxy model at t=tinit assuming a given star fromation model and cosmology
    g = gm.model_galaxy(t = t_output[0], Mh = Minit, Mg = None, Ms = None, MZ = None, 
                     sfrmodel = sfrmodel, cosmo = cosmo, verbose = verbose)
    
    # array of quantities to evolve with ODE solver
    y0 = np.array([g.Mh, g.Mg, g.Ms, g.MZ])
    # evolve y0 in time
    Mout = odeint(g.evolve, y0, t_output, rtol = 1.e-7, mxstep = 4000)
    
    # let's split the output into arrays with intuitive names
    Mhout = np.clip(Mout[:,0], 0.0, 1.e100)
    Mgout = np.clip(Mout[:,1], 0.0, 1.e100)
    Msout = np.clip(Mout[:,2], 0.0, 1.e100)
    MZout = np.clip(Mout[:,3], 0.0, 1.e100)

    return zg, t_output, Mhout, Mgout, Msout, MZout
```

```python
sfrmodel = "gaslinear"  
# let's take a look at how a halo that has MW-like halo mass at z=0 evolves
# the class method test_galaxy_evolution does some basic run of the model and plots results, so let's setup 
# matplotlib parameters for plotting
from code.plot_utils import plot_pretty
%matplotlib inline
from matplotlib import pylab as plt
plot_pretty()

zg, t_output, Mhout, Mgout, Msout, MZout = test_galaxy_evolution(Minit=2.1e8, 
                                                                 sfrmodel=sfrmodel, cosmo=cosmo, 
                                                                 verbose=False)

#
# plot
#
tu = cosmo.age(0.0) # the age of the Universe

fig, ax = plt.subplots(1,2, figsize=(6,3))    
ax[0].set_xlabel(r'$t\ \rm (Gyr)$'); ax[0].set_ylabel(r'$M_{\rm h},\ M_{\rm g},\ M_{*},\ M_{\rm Z}\ (\rm M_\odot)$')
ax[0].set_xlim(0.,tu); ax[0].set_ylim(1.e4,1.e13)
ax[0].set_yscale('log')
ax[0].plot(t_output, Mhout, lw = 3.0, label=r'$M_{\rm vir}$')
ax[0].plot(t_output, Mgout, lw = 3.0, label=r'$M_{\rm gas}$')
ax[0].plot(t_output, Msout, lw = 3.0, label=r'$M_{\rm *}$')
ax[0].plot(t_output, MZout, lw = 3.0, label=r'$M_{\rm Z}$')

ax[0].legend(frameon=False,loc='lower right', ncol=2, fontsize=9)
ax[0].grid(linestyle='dotted', c='lightgray')
    
# plot SFR and Z(t)
ax[1].set_xlabel(r'$t\ \rm (Gyr)$'); ax[1].set_ylabel(r'$Z= M_{\rm Z}/M_{\rm g}\ (Z_\odot),\rm\ SFR\ (M_\odot/yr)$')
ax[1].set_xlim(0.,tu); ax[1].set_ylim(1.e-5,30.)
ax[1].set_yscale('log')
ax[1].yaxis.set_label_position('right')


ax[1].plot(t_output, MZout/Mgout/0.02, lw = 3.0, label=r'$Z$')
SFR = np.zeros_like(t_output); 

for i, td in enumerate(t_output):
    ge = gm.model_galaxy(t = td, Mh = Mhout[i], Mg = Mgout[i], Ms = Msout[i], MZ = MZout[i], 
                         sfrmodel = sfrmodel, cosmo = cosmo)
    SFR[i] = ge.SFR(td)*1.e-9;

ax[1].plot(t_output,SFR, lw = 3.0, label=r'$\rm SFR$')
ax[1].legend(frameon=False,loc='lower right', ncol=2, fontsize=9)
ax[1].grid(linestyle='dotted', c='lightgray')
plt.show()
```

##### Solution

As we can see the total halo mass monotonically incereases. This can be understood in terms of peak collapse process. Collapse happens when an overdensity / peak reaches a critical value (~ that calculated using the spherical collapse model). As time progresses more overdense regions reach that critical value virialize; thus the virial mass increases. 

Since, as we have seen, there is a relation between halo mass and total stellar mass, it is reasonable to expect a similar behavior of the stellar mass as a function of time.

On the other hand, the total mass in gas decreases as it condenses in stars.

```python
from scipy.interpolate import interp2d

# initial metallicity
Z_IGM = 2.e-5

# grid of redshifts and halo masses
Nz = 41; Nm = 100
zg = np.linspace(30., 0., Nz)
lzg1 = np.log10(1.0+zg)
t_output = cosmo.age(zg)
    
lMh = np.linspace(5.0, 8., Nm)
Mh  = np.power(10, lMh)

# arrays to hold evolutionary grids of halo mass, gas mass, stellar mass and metal mass
Mhout = np.zeros((Nm,Nz)); Mgout = np.zeros((Nm,Nz))
Msout = np.zeros((Nm,Nz)); MZout = np.zeros((Nm,Nz))
#evolve a grid of halo masses
for j, Mhd in enumerate(Mh):
    if not j%5:
        print "evolving halo of log10(Mh)=%2.2f ..."%(lMh[j])
    # instantiate a model galaxy with a given halo mass at the initial time
    g = gm.model_galaxy(t = t_output[0], Mh = Mhd, Mg = None, Ms = None, MZ = None, Z_IGM = Z_IGM, sfrmodel = sfrmodel, cosmo = cosmo)
    # initial values of masses
    y0 = np.array([g.Mh, g.Mg, g.Ms, g.MZ])
    # solve the system of ODEs
    Mout = odeint(g.evolve, y0, t_output, rtol = 1.e-5, mxstep = 4000)
    Mout = Mout/g.cosmo.h
    # split Mout into arrays for specific masses with more intuitive names for convenience
    Mhout[j,:] = Mout[:,0]; Mgout[j,:] = Mout[:,1]
    Msout[j,:] = Mout[:,2]; MZout[j,:] = Mout[:,3]
# 
# prepare 2D splines for interpolation
#
Mmin = 1.e-10 # prevent zeros in case no SF occurred
lMhi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Mhout)), bounds_error=True, kind='cubic')
lMgi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Mgout)), bounds_error=True, kind='cubic')
lMsi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Msout)), bounds_error=True, kind='cubic')
lMZi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,MZout)), bounds_error=True, kind='cubic')
```

```python
#
# plot
#

fig = plt.figure(figsize=(3, 3))
plot_pretty()
plt.xlabel(r'$M_h\ \rm (M_\odot)$')
plt.ylabel(r'$M_{*}\ \rm (M_\odot)$')
plt.xlim(1.e7,1.e15); plt.ylim(1.e4,1.e13)
plt.xscale('log'); plt.yscale('log')
        
z_plot = np.array([0., 2.2, 3.5, 6.0])
# colors a la Feldmann for easier comparison
zcolors = ['green', 'red', 'blue', 'darkgray']
for i, zplot in enumerate(z_plot):
    Mhz = 10.**lMhi(np.log10(1.0+zplot),lMh)
    Msz = 10.**lMsi(np.log10(1.0+zplot),lMh)
    plt.plot(Mhz, Msz, c=zcolors[i], lw=3.0, label=r'$z=%1.1f$'%zplot)
            
plt.legend(frameon=False,loc='lower right', fontsize=10)
plt.grid(linestyle='dotted', c='lightgray')
plt.show()
```
